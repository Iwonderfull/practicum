## Предобработка данных

### Парсинг даты

```python
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'],format='%Y-%m-%dT%H:%M:%S')
```

```python
df['year_exposition'] = df['first_day_exposition'].dt.year 
df['month_exposition'] = df['first_day_exposition'].dt.month
df['day_exposition'] = df['first_day_exposition'].dt.day
df['weekday_exposition'] = df['first_day_exposition'].dt.weekday #strftime('%a')

df.rename(
    columns={
        'first_day_exposition': 'date_exposition', 
        'days_exposition': 'n_days_exposition'
    }, 
    inplace=True)
```

```python
df[['date_exposition', 'year_exposition', 'month_exposition', 'day_exposition', 'weekday_exposition']].head(5)
```

Даты оформатированы. 

Добавлены колонки ['year_exposition', 'month_exposition', 'day_exposition', 'weekday_exposition'] для года, месяца, дня месяца и дня недели выставления объявления о продаже соответственно.

Колонка 'first_day_exposition' переименована в 'date_exposition'

### Восстановление площадей

```python
test = df.loc[df[['living_area','kitchen_area', 'total_area']].notna().all(axis=1)]
(test['total_area'] - (test['living_area'] + test['kitchen_area'])).hist()
```

Тогда подсчитаем соотношение жилой/общей и кухонной/общей, найдем медианы и с их помощью заполним пропуски. Отделять при этом выбросы в площадях не будем т.к. медиана к ним устойчива.

```python
# Add area ratios
df['kitchen_total_ratio'] = df['kitchen_area'] / df['total_area']
df['living_total_ratio'] = df['living_area'] / df['total_area']

df[['kitchen_total_ratio', 'living_total_ratio']].boxplot()
print(df[['kitchen_total_ratio', 'living_total_ratio']].median())
```

Получается, что площадь кухни в среднем 17%, а жилой - 57% от общей площади, не смотря на выбросы. 

Заполним пропуски с помощью полученных соотношений.

```python
# Fill NA
df['living_area'].fillna(df['living_total_ratio'].median() * df['total_area'], inplace=True)
df['kitchen_area'].fillna(df['kitchen_total_ratio'].median() * df['total_area'], inplace=True)
```

```python
print(
    "Количество пропусков в значениях 'living_area','kitchen_area':",
    df[['living_area','kitchen_area']].isna().any(axis=1).sum()
)
```

### Восстановление количества дней размещения

Посмотрим где находятся пропуски 'days_exposition' относительно года и месяца публикации объявления.

```python
# Create month abbriviations list
month_abbs = pd.date_range('2021-01-01','2021-12-01', freq='MS').strftime("%b").tolist()
```

```python
# Pivot table by year and month where days_exposition is NA
na_exposition = df.query('n_days_exposition.isna()').pivot_table(
    index=['year_exposition'], 
    columns=['month_exposition'], 
    values=['day_exposition'],
    aggfunc='count',
    fill_value=0
).applymap(int)

# Change column names by month abbreviations
na_exposition.columns = month_abbs

# Plot seaborn heatmap 
sns.heatmap(
    na_exposition, 
    annot=True, 
    fmt="4", 
    linewidths=.5
).set_title("Dates where days_exposition DOESN'T exists" )

print('Значения в ячейках - количество пропусков')
```

Большинство пропущенных значений приходятся на 2018, 2019 годы и больше всего их в апреле 2019. При чем наблюдается тенденция роста числа пропусков 'days_exposition' с приближением даты к настоящему моменту. По всей видимости эти объявления еще не сняты (чтобы был спрос должено же быть предложение). Но с 2014 по 2017 пусть будет частью объявлений, которую забыли снять. 

**Надо ли оставлять все?**

Теперь посмотрим на количество указанных значений 'days_exposition' по годам и месяцам

```python
# Pivot table by year and month where days_exposition exists
exists_exposition = df.query('n_days_exposition.notna()').pivot_table(
    index=['year_exposition'], 
    columns=['month_exposition'], 
    values=['day_exposition'],
    aggfunc='count',
    fill_value=0
).applymap(int)

# Change column names by month abbreviations
exists_exposition.columns = month_abbs

# Plot seaborn heatmap 
sns.heatmap(
    exists_exposition, 
    annot=True, 
    fmt="4", 
    linewidths=.5).set_title('Dates with EXISTS days_exposition')

print('Значения в ячейках - количество заполненных значений')
```

В сводных таблицах пропущенные занчения и указанные значения находятся в одних и тех же группах. Т.е. если для существующей группы (год, месяц) нет указанных значений, то и пропущенных тоже нет. 

Заменим пропуски в 'days_exposition' с 2014 по 2017 включительно на медианы по году и месяцу из указанных значений. При чем значения будем брать из тех же временных групп. В некоторых случаях можно было бы заменить на медианы от прошлого года, но тогда в нашем случае количество дней размещения объявления в среднем было бы больше на год. 

**Как оценить вносимую погрешность, т.е. не исказим ли мы данные такими переменами?**

```python
# Pivot table with 'days_exposition' medians by date
days_exposition_medians = df.query('n_days_exposition.notna()').pivot_table(
    index=['year_exposition', 'month_exposition'], 
    values='n_days_exposition', 
    aggfunc='median'
)

# Uncomment to see them. Pivot table or dict of it.
# days_exposition_medians
# days_exposition_medians['days_exposition'].to_dict()
```

```python
n_days_exposition_miss = df['n_days_exposition'].isna() & (~df['year_exposition'].isin([2019, 2018]))

for idx, value in days_exposition_medians['n_days_exposition'].to_dict().items():
    
    df.loc[(n_days_exposition_miss & 
            (df['year_exposition']==idx[0]) & 
            (df['month_exposition'] == idx[1])), 
           'n_days_exposition'] = value


df.loc[:, ['year_exposition', 'month_exposition', 'n_days_exposition']].head(10)
```

Пропуски в 'days_exposition' частично заполнены медианами указанных значений по году и месяцу.

```python
df.pivot_table(
    index='year_exposition', 
    columns='balcony', 
    values='n_days_exposition', 
    aggfunc='count', 
    fill_value=0)
```

Таким образом в объявлениях с 2014 по 2017 годы самые популярные квартиры без балконов, а с 2018 с одним балконом.

В колонке с количеством балконов пропуски восстановлены модой относительно года публикации объявления.
