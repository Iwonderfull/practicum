### Расчет rougher.recovery

<img src="https://pictures.s3.yandex.net/resources/Recovery_1576238822.jpg" />

На всякий случай в знаменатели добавил маленькую епсилон, чтобы уклониться от деления на ноль.

```python
def recovery(inputs, concentrate, tails):
    left = concentrate / (inputs + 1e-25)
    right = (inputs - tails) / (concentrate - tails + 1e-25)
    return left * right * 100 
```

```python
data_train_rougher_recovery = recovery(
    inputs = data_train["rougher.input.feed_au"], 
    concentrate=data_train["rougher.output.concentrate_au"], 
    tails=data_train["rougher.output.tail_au"]
)
```

```python
subtraction_error = data_train_rougher_recovery - data_train["rougher.output.recovery"]

ax = sns.histplot(subtraction_error)
ax.set_title("Распределение разностей расчетов и recovery данных")
ax.set_xlabel("Значение разности")
ax.set_ylabel("Количество")

subtraction_error.describe()
```

```python
train_rougher_recovery_mae = mean_absolute_error(
    data_train_rougher_recovery, 
    data_train['rougher.output.recovery']
)
print("Train rougher recovery MAE: {:e}".format(train_rougher_recovery_mae)) 
```

Расчетный коэффициент recovery и коэффициент в данных для выборки train имеют небольшое расхождение, порядка 1.12e-14 по метрике MAE. Их разность имеет распределение похожее на нормальное с центром в нуле и небольшим стандартным отклонением в 1.6e-14% (т.к. эффективность обогащения в процентах). Положим за незначительную погрешность измерений.

### Удаление выбросов

```python
def where_zeros(df, columns):
    idx_with_zeros = df[(df[columns] ==0 ).any(axis=1)].index
    print("Количество объектов с нулями:", len(idx_with_zeros))
    print("Доля объектов с нулями: {:.0%}".format(len(idx_with_zeros) / len(df)))
    return idx_with_zeros
```

```python
print("Удалено")
data_full = data_full.drop(
    index=where_zeros(data_full, metal_cols)
)
```

## Модель

### sMAPE

<img src="https://pictures.s3.yandex.net/resources/smape_1576238825.jpg" />

```python
def sMAPE(y_true, y_pred):
    error = np.abs(y_true - y_pred) 
    scale = (np.abs(y_true) + np.abs(y_pred)) / 2
    return np.mean(error / (scale + 1e-30)) * 100
```

<img src="https://pictures.s3.yandex.net/resources/_smape_1576238814.jpg"/>

```python
def final_sMAPE(y_true, y_pred, weights=[.25, .75]):
    error = sMAPE(y_true, y_pred) * np.array(weights)
    return np.sum(error)
```

### Подкготовка выборок

```python
# X_train = data_train[data_test.columns]
# X_test = data_test

# y_rougher_train = data_train['rougher.output.recovery']
# y_final_train = data_train['final.output.recovery']

# y_rougher_test = data_full.loc[data_test.index, 'rougher.output.recovery']
# y_final_test = data_full.loc[data_test.index, 'final.output.recovery']

X_train = data_train[data_test.columns]
X_test = data_test

y_train = data_train[['rougher.output.recovery', 'final.output.recovery']]

y_test = data_full.loc[data_test.index, ['rougher.output.recovery', 'final.output.recovery']]
```

### Модель

```python
rougher_model = RandomForestRegressor(random_state=12345)
final_model = RandomForestRegressor(random_state=12345)
# param_grid = ParameterGrid(
#    {'rougher__max_depth': [3,5,10],
#     'rougher__n_estimators': [10, 50, 100],
#     'final__max_depth': [3,5,10],
#     'final__n_estimators': [10, 50, 100]}
# )
param_grid = ParameterSampler(
   {'rougher__max_depth': st.randint(3, 15),
    'rougher__n_estimators': st.randint(3, 100),
    'final__max_depth': st.randint(3, 15),
    'final__n_estimators': st.randint(3, 100)},
    n_iter=10,
    random_state=12345
)
cv_iter = ShuffleSplit(n_splits=5, test_size=.2, random_state=12345)
```

### Кроссвалидация

```python
best_params = None
best_score = None

rougher_scores=[]
final_scores=[]
scores=[]
# Iterate over parameters grid
for step, params in enumerate(param_grid):
    
    # Set new model params
    rougher_model.set_params(
        max_depth=params['rougher__max_depth'],
        n_estimators=params['rougher__n_estimators']
    )
    final_model.set_params(
        max_depth=params['final__max_depth'],
        n_estimators=params['final__n_estimators']
    )
    
    # Flush CV scores
    cv_rougher_scores=[]
    cv_final_scores=[]
    cv_scores=[]
    
    # Cross validation
    for cv_train_idx, cv_val_idx in cv_iter.split(X=X_train):
        
        # Fit models
        rougher_model.fit(
            X_train.iloc[cv_train_idx], 
            y_train.iloc[cv_train_idx, 0]
        )
            
        final_model.fit(
            X_train.iloc[cv_train_idx], 
            y_train.iloc[cv_train_idx, 1]
        )
        
        #Get CV predictions
        cv_rougher_predictions = rougher_model.predict(X_train.iloc[cv_val_idx])
        cv_final_predictions = final_model.predict(X_train.iloc[cv_val_idx])
        
        # Get CV sMAPE scores for each model
        cv_rougher_scores.append(
            sMAPE(
                y_train.iloc[cv_val_idx, 0],
                cv_rougher_predictions
            )
        )
        cv_final_scores.append(
             sMAPE(
                 y_train.iloc[cv_val_idx, 1],
                 cv_final_predictions
             )
        )
        # Get CV combined sMAPE score
        cv_scores.append(
            final_sMAPE(
                y_train.iloc[cv_val_idx], 
                np.stack([cv_rougher_predictions,cv_final_predictions], axis=1)
            )
        )
    # Get mean CV scores 
    rougher_scores.append(np.mean(cv_rougher_scores))
    final_scores.append(np.mean(cv_final_scores))
    scores.append(np.mean(cv_scores))
    
    # Save best score/params
    if best_score is None or scores[-1] < best_score:
        best_score = scores[-1]
        best_params = params
    
    # Print results
    print("Params grid step [{}/{}]:".format(step + 1, len(param_grid)))
    print("    rougher stage sMAPE: {:.2f}".format(rougher_scores[-1]))
    print("    final stage sMAPE: {:.2f}".format(final_scores[-1]))
    print("    sMAPE: {:.2f}".format(scores[-1]))
    print("    params:")
    for k,v in params.items():
        print("         {!r}:{}".format(k, v))
# Print best score/params
print()
print("Best combined sMAPE score: {:.2f}".format(best_score))
print("Best params:")
for k,v in best_params.items():
    print("    {!r}:{}".format(k, v))
```

```python
# Refit models with best params
rougher_model.set_params(
    max_depth=best_params['rougher__max_depth'],
    n_estimators=best_params['rougher__n_estimators']
)
final_model.set_params(
    max_depth=best_params['final__max_depth'],
    n_estimators=best_params['final__n_estimators']
)
for cv_train_idx, cv_val_idx in cv_iter.split(X=X_train):
        # Fit models
        rougher_model.fit(
            X_train.iloc[cv_train_idx], 
            y_train.iloc[cv_train_idx, 0]
        )
            
        final_model.fit(
            X_train.iloc[cv_train_idx], 
            y_train.iloc[cv_train_idx, 1]
        )
```

### sMAPE на тестовой выборке

```python
# Test score calculation
#Get test predictions
test_rougher_predictions = rougher_model.predict(X_test)
test_final_predictions = final_model.predict(X_test)

# Get test sMAPE scores for each model
test_rougher_scores = sMAPE(
    y_test.iloc[:, 0],
    test_rougher_predictions
)

test_final_scores = sMAPE(
     y_test.iloc[:, 1],
     test_final_predictions
)

# Get test combined sMAPE score
test_score = final_sMAPE(
    y_test,
    np.stack([test_rougher_predictions, test_final_predictions], axis=1)
)
```
